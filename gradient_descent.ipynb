```python
import matplotlib.pyplot as plt

import numpy as np
# import the data from the the data01.txt file
data = np.loadtxt('data01.txt', delimiter='\t')

# split the data into x and y
x = data[:, 0]
y = data[:, 1]

# plot the data

plt.scatter(x, y, marker='o', c='b',s=2)

plt.show()
```

```python
# y = mx + c
#inital values of m and c
m=0
c=0 

#learning rate
L=0.0001

#number of iterations
epochs=100

#cost function
n = float(len(x))
#perform gradient descent

for i in range(epochs):
    y_predicted = m * x + c #predicted y value

    j = (1/n) * sum([val**2 for val in (y-y_predicted)]) #cost function


    # plt.scatter(m,j, marker='o', c='b',s=2)
    # plt.scatter(c,j ,marker='o', c='r',s=2)

    # subplot m,j and c,j
    plt.subplot(1,2,1)
    plt.scatter(m,j, marker='o', c='b',s=2)
    plt.xlabel('m')
    plt.ylabel('j')
    plt.subplot(1,2,2)
    plt.scatter(c,j ,marker='o', c='r',s=2)
    plt.xlabel('c')
    plt.ylabel('j')
    # plt.pause(0.05)
    # plt.clf()


    #get gradients of m and c

    m_gradient = -(2/n) * sum(x * (y - y_predicted))
    c_gradient = -(2/n) * sum(y - y_predicted)

    #update m and c

    m = m - L * m_gradient
    c = c - L * c_gradient


plt.pause(0.05)
plt.clf()
print(m,c)
y_predicted = m * x + c
plt.plot(x, y_predicted, c='g')
plt.scatter(x, y, marker='o', c='b',s=2)

plt.show()   
```
